```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(colombianos,iteraciones):
    contador = 0.
    for i in range(iteraciones):
        nums = np.ones(10.+colombianos)
        nums[:10] = 0.
        np.random.shuffle(nums)
            
        for j in range(len(nums)-1):
            if(nums[j]==0 and nums[j-1]==1. and nums[j+1]==1.):
                contador += 1.
    return contador/10./iteraciones

def f_teo(colombianos):
    return colombianos*(colombianos-1.)/(colombianos+10.-1.)/(colombianos+10.-2.)
```

```python
x = np.arange(2,60)
y_0 = f_teo(x)
y = np.array([f(t,1000) for t in x])
plt.figure(figsize=(15,10))
plt.plot(x,y_0,'r',ls='dashed',label=u'Analítica')
plt.plot(x,y,'bo',label=u'Empírica')
plt.xlabel('Colombianos')
plt.ylabel('Probabilidad de estar rodeado')
plt.legend(loc='best')
plt.show()
```
