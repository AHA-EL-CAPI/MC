```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import matplotlib.animation as animation
```

```python
%matplotlib inline
alpha = -15.
times=np.linspace(0,1.,100000)
def f(Y,t):
    return np.array([Y[1],-alpha*np.cos(Y[0])*Y[1]]) 


funsol,others=odeint(f,[0.,1.],times,full_output=True)

plt.figure(figsize=(15,10))
plt.plot(times,funsol[:,0],'k')
plt.plot(times,funsol[:,1],'b')
plt.show()
```

```python
%matplotlib
fig = plt.figure(figsize=(15,10))
times=np.linspace(0,1.,100000)
pos, = plt.plot([],[],'k')
vel, = plt.plot([],[],'b')
plt.xlim(0.,1.)

alphas = np.arange(-15.,16.)
def animate(i):
    def f(Y,t):
        return np.array([Y[1],-alphas[i]*np.cos(Y[0])*Y[1]]) 
    funsol,others=odeint(f,[0.,1.],times,full_output=True)
    pos.set_data(times,funsol[:,0])
    vel.set_data(times,funsol[:,1])
    plt.ylim(0.,max(funsol[:,1])+0.1)
    return pos,vel

anim = animation.FuncAnimation(fig,animate,30,interval=500,blit=False,repeat_delay=1000)
plt.show()
```

```python
%matplotlib inline
mass = 1.
omega = 1.
alpha = -1.
def f1(Y,t):
    return np.array([Y[1],-alpha*np.cos(Y[0])*Y[1] - (omega**2)*Y[0] + (1/mass)*np.cos(Y[0])*np.exp(t)]) 

def f2(Y,t):
    return np.array([Y[1],-alpha*Y[1] - (omega**2)*Y[0] + (1/mass)*np.cos(Y[0])*np.exp(t)]) 

times=np.linspace(0,15.,100000)
funsol1,others1=odeint(f1,[0.,1.],times,full_output=True)
funsol2,others2=odeint(f2,[0.,1.],times,full_output=True)

plt.figure(figsize=(15,10))
plt.plot(times,funsol1[:,0],'r')
plt.plot(times,funsol2[:,0],'b')
#plt.savefig('Morales es marica')
plt.show()
```

```python
%matplotlib
fig = plt.figure(figsize=(15,10))
times=np.linspace(0,15.,100000)
mass = 1.
omega = 1.
pos, = plt.plot([],[],'k')
vel, = plt.plot([],[],'b')
plt.xlim(0.,15.)

alphas = np.linspace(-.4,.4,100)
def animate(i):
    def f1(Y,t):
        return np.array([Y[1],-alphas[i]*np.cos(Y[0])*Y[1] - (omega**2)*Y[0] + (1./mass)*np.cos(Y[0])*np.exp(t)]) 

    def f2(Y,t):
        return np.array([Y[1],-alphas[i]*Y[1] - (omega**2)*Y[0] + (1./mass)*np.cos(Y[0])*np.exp(t)]) 

    funsol1,others1=odeint(f1,[0.,1.],times,full_output=True)
    funsol2,others2=odeint(f2,[0.,1.],times,full_output=True)
    pos.set_data(times,funsol1[:,0])
    vel.set_data(times,funsol2[:,0])
    plt.ylim(min(funsol2[:,0])-0.1,max(funsol2[:,0])+0.1)
    return pos,vel

anim = animation.FuncAnimation(fig,animate,100,interval=1,blit=False,repeat_delay=1000)
plt.show()
```
