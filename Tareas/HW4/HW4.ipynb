```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 2**10 # Número de puntos de muestreo
a = np.zeros((N,N))

# Genero la rendija cuadrada de lado 32 usando slice notation
a[496:528,496:528] = 1

b = np.fft.fft2(a) # Hago la Transformada de Fourier Discreta (DFT) teniendo en cuenta que tengo una array bidimensional.

b = np.fft.fftshift(b) # Aplico la función shift al resultado de la DFT.

b = np.abs(b)**2 +1 # Obtengo el módulo al cuadrado del array anterior. sumo 1 para evitar problemas al resaltar con log.

b = np.log(b) # Tomo el logaritmo del módulo al cuadrado y sumo 1 por el problema mencionado en el enunciado.

# Obtengo información acerca de un cuadrado de lado 200 en el arreglo  anterior de acuerdo a lo solicitado en el enunciado.
b = b[412:612,412:612]

plt.figure(figsize=(20,20))        
plt.subplot(1,2,1,aspect='equal')
plt.imshow(a,cmap='Greys_r')
plt.subplot(1,2,2,aspect='equal')
plt.title(u'Patrón de interferencia de una apertura cuadrada \n (log resaltado)')
plt.imshow(b)
plt.show()
```

```python
N = 2**10 # Número de puntos de muestreo
a = np.zeros((N,N))

# Genero la doble rendija
a[412:612,450:470] = 1
a[412:612,554:574] = 1
    
b = np.fft.fft2(a) # Hago la Transformada de Fourier Discreta (DFT) teniendo en cuenta que tengo una array bidimensional.

b = np.fft.fftshift(b) # Aplico la función shift al resultado de la DFT.

b = np.abs(b)**2 # Tomo el módulo al cuadrado.

# Obtengo información acerca de un cuadrado de lado 128 en el arreglo  anterior de acuerdo a lo solicitado en el enunciado.
b = b[448:576,448:576]
        
# Grafico los arreglos obtenidos previamente

plt.figure(figsize=(20,20))        
plt.subplot(1,2,1,aspect='equal')
plt.imshow(a,cmap='Greys_r')
plt.subplot(1,2,2,aspect='equal')
plt.title(u'Patrón de difracción de una doble rendija',fontsize=20)
plt.imshow(b)
plt.show()
```

```python
N = 2**10 # Número de puntos de muestreo
radius = 10.24 # Radio de la apertura teniendo en cuenta que es una centésima parte de la longitud de la placa
               # que en ete caso es 1024

cx, cy = 512., 512. # Defino el centro del círculo

a = np.array([[(((x-cx)**2+(y-cy)**2) <= radius**2) for x in range(N)] for y in range(N)])

b = np.fft.fft2(a) # Hago la Transformada de Fourier Discreta (DFT) teniendo en cuenta que tengo una array bidimensional.

b = np.fft.fftshift(b) # Aplico la función shift al resultado de la DFT.

b = np.abs(b)**2 +1

c = np.log(b) # Tomo el logaritmo del módulo al cuadrado y sumo 1 por el problema mencionado en el enunciado.

# Grafico los arreglos obtenidos previamente usando las ayudas del enunciado.

plt.figure(figsize=(20,20)) 

plt.subplot2grid((3,2),(0,0),colspan=2,rowspan=1)
x = (2.*np.pi*radius/1024.)*np.linspace(-512,512,1024)
plt.plot(x,b[512]/max(b[512]),'bo')
plt.xlim(0.,10.)
plt.xlabel(r'$2\pi$'+'$Rj/L$',fontsize=20)
plt.ylabel('$I/I_0$',fontsize=20)

plt.subplot2grid((3,2),(1,0),rowspan=2)
plt.imshow(a,cmap='Greys_r')

plt.subplot2grid((3,2),(1,1),rowspan=2)
plt.title(u'Patrón de difracción de una apertura circular (Patrón de Airy) \n log resaltado',fontsize=20)
plt.imshow(c)

plt.tight_layout()
plt.show()
```
