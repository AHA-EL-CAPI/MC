```python
import numpy as np
import matplotlib.pyplot as plt
import itertools
%matplotlib inline
```

```python
N = 2**10 # Número de puntos de muestreo
a = np.zeros((N,N))

# Genero la rendija cuadrada
for i in range(496,528,1):
    a[i][496:528] = 1
    
b = np.fft.fft2(a) # Hago la Transformada de Fourier Discreta (DFT) teniendo en cuenta que tengo una array bidimensional.

b = np.fft.fftshift(b) # Aplico la función shift al resultado de la DFT.

b = np.log(b**2 +1) # Tomo el logaritmo del módulo al cuadrado y sumo 1 por el problema mencionado en el enunciado.

# Genero y grafico un arreglo con la información acerca de un cuadrado de lado 200 en el arreglo que contiene la
# información sobre la DFT y el shift
c = np.zeros((200,200))

for i in range(200):
    for j in range(200):
        c[i][j] = b[412+i][412+j].real 
        
# Grafico los arreglos obtenidos previamente

plt.figure(figsize=(20,20))        
plt.subplot(1,2,1,aspect='equal')
plt.imshow(a,cmap='Greys_r')
plt.subplot(1,2,2,aspect='equal')
plt.title(u'Patrón de interferencia de una apertura cuadrada \n (log resaltado)')
plt.imshow(c)
plt.show()
```

```python
N = 2**10 # Número de puntos de muestreo
a = np.zeros((N,N))

# Genero la doble rendija
for i in range(412,612,1):
    a[i][450:470] = 1
    a[i][554:574] = 1
    
b = np.fft.fft2(a) # Hago la Transformada de Fourier Discreta (DFT) teniendo en cuenta que tengo una array bidimensional.

b = np.fft.fftshift(b) # Aplico la función shift al resultado de la DFT.

b = b**2 # Tomo el módulo al cuadrado.

# Genero y grafico un arreglo con la información acerca de un cuadrado de lado 200 en el arreglo que contiene la
# información sobre la DFT y el shift
c = np.zeros((128,128))

for i in range(128):
    for j in range(128):
        c[i][j] = np.abs(b[448+i][448+j]) 
        
# Grafico los arreglos obtenidos previamente

plt.figure(figsize=(20,20))        
plt.subplot(1,2,1,aspect='equal')
plt.imshow(a,cmap='Greys_r')
plt.subplot(1,2,2,aspect='equal')
plt.title(u'Patrón de difracción de una doble rendija')
plt.imshow(c)
plt.show()
```

```python
radius = 10
a = np.zeros((N, N))
cx, cy = 512, 512 # The center of circle
y, x = np.ogrid[-radius: radius, -radius: radius]
index = x**2 + y**2 <= radius**2
a[cy-radius:cy+radius, cx-radius:cx+radius][index] = 255

b = np.fft.fft2(a) # Hago la Transformada de Fourier Discreta (DFT) teniendo en cuenta que tengo una array bidimensional.

b = np.fft.fftshift(b) # Aplico la función shift al resultado de la DFT.

c = np.log(b**2 +1) # Tomo el módulo al cuadrado.

# Grafico los arreglos obtenidos previamente

plt.figure(figsize=(20,20))        
plt.subplot2grid((3,2),(0,0),colspan=2,rowspan=1)
x = 2.*np.pi*radius/1000.*np.linspace(-512,512,1024)
plt.plot(x,b[512].real,'bo')
plt.subplot2grid((3,2),(1,0),rowspan=2)
plt.imshow(a,cmap='Greys_r')
plt.subplot2grid((3,2),(1,1),rowspan=2)
plt.title(u'Patrón de difracción de una apertura circular (Patrón de Airy) \n log resaltado')
plt.imshow(np.abs(c))
plt.tight_layout()
plt.show()
```
