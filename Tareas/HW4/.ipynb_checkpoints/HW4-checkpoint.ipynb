```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 2**10 # Número de puntos de muestreo
a = np.zeros((N,N))

# Genero la rendija cuadrada de lado 32 usando slice notation
a[496:528,496:528] = 1

b = np.fft.fft2(a) # Hago la Transformada de Fourier Discreta (DFT) teniendo en cuenta que tengo una array bidimensional.

b = np.fft.fftshift(b) # Aplico la función shift al resultado de la DFT.

b = np.abs(b)**2 +1 # Obtengo el módulo al cuadrado del array anterior. sumo 1 para evitar problemas al resaltar con log.

b = np.log(b) # Tomo el logaritmo del módulo al cuadrado y sumo 1 por el problema mencionado en el enunciado.

# Obtengo información acerca de un cuadrado de lado 200 en el arreglo  anterior de acuerdo a lo solicitado en el enunciado.
b = b[412:612,412:612]

plt.figure(figsize=(20,20))        
plt.subplot(1,2,1,aspect='equal')
plt.imshow(a,cmap='Greys_r')
plt.subplot(1,2,2,aspect='equal')
plt.title(u'Patrón de interferencia de una apertura cuadrada \n (log resaltado)',fontsize=18)
plt.imshow(b)
plt.show()
```

```python
N = 2**10 # Número de puntos de muestreo
a = np.zeros((N,N))

# Genero la doble rendija
a[412:612,450:470] = 1
a[412:612,554:574] = 1
    
b = np.fft.fft2(a) # Hago la Transformada de Fourier Discreta (DFT) teniendo en cuenta que tengo una array bidimensional.

b = np.fft.fftshift(b) # Aplico la función shift al resultado de la DFT.

b = np.abs(b)**2 # Tomo el módulo al cuadrado.

# Obtengo información acerca de un cuadrado de lado 128 en el arreglo  anterior de acuerdo a lo solicitado en el enunciado.
b = b[448:576,448:576]
        
# Grafico los arreglos obtenidos previamente

plt.figure(figsize=(20,20))        
plt.subplot(1,2,1,aspect='equal')
plt.imshow(a,cmap='Greys_r')
plt.subplot(1,2,2,aspect='equal')
plt.title(u'Patrón de difracción de una doble rendija',fontsize=20)
plt.imshow(b)
plt.show()
```

```python
N = 2**10 # Número de puntos de muestreo
radius = 10.24 # Radio de la apertura teniendo en cuenta que es una centésima parte de la longitud de la placa
               # que en ete caso es 1024

cx, cy = 512., 512. # Defino el centro del círculo

a = np.array([[(((x-cx)**2+(y-cy)**2) <= radius**2) for x in range(N)] for y in range(N)]) 

# Para contruir el arreglo anterior se hizo primero con ceros y unos, pero la gráfica que muestra el paatrón de Airy 
# resaltado no tenía buena resolución. Por este motivo, buscamos cómo hacer mejor la gráfica y se obtuvo el código 
# anterior del github de Diva.


b = np.fft.fft2(a) # Hago la Transformada de Fourier Discreta (DFT) teniendo en cuenta que tengo una array bidimensional.

b = np.fft.fftshift(b) # Aplico la función shift al resultado de la DFT.

b1 = np.abs(b)**2 +1

c = np.log(b1) # Tomo el logaritmo del módulo al cuadrado y sumo 1 por el problema mencionado en el enunciado.

# Grafico los arreglos obtenidos previamente usando las ayudas del enunciado.

plt.figure(figsize=(20,20)) 

plt.subplot2grid((3,2),(0,0),colspan=2,rowspan=1)
x = (2.*np.pi*radius/1024.)*np.linspace(-512,512,1024)
plt.plot(x,b1[512]/max(b1[512]),'bo')
plt.xlim(0.,10.)
plt.xlabel(r'$2\pi$'+'$Rj/L$',fontsize=20)
plt.ylabel('$I/I_0$',fontsize=20)

plt.subplot2grid((3,2),(1,0),rowspan=2)
plt.imshow(a,cmap='Greys_r')

plt.subplot2grid((3,2),(1,1),rowspan=2)
plt.title(u'Patrón de difracción de una apertura circular (Patrón de Airy) \n log resaltado',fontsize=20)
plt.imshow(c)

plt.tight_layout()
plt.show()
```

```python
# Interpolación Cúbica. Basado en el taller 4 esta era una de las mejores opciones 
# para interpolar. Se intentó realizar una interpolación de Lagrange pero tardaba mucho tiempo.
from scipy import interpolate

radius = 10.24
x = (2.*np.pi*radius/1024.)*np.linspace(-512,512,1024)
y = b1[512]/max(b1[512])

inter = interpolate.interp1d(x,y,kind='cubic')
```

```python
from scipy.signal import argrelextrema 
# Se importa esta herramienta para buscar los máximos y mínimos locales deseados. La información se obtuvo de la página:
# http://stackoverflow.com/questions/4624970/finding-local-maxima-minima-with-numpy-in-a-1d-numpy-array

from tabulate import tabulate # Herramienta necesaria para construir las tablas mostradas al final.

loc_max = argrelextrema(inter(x), np.greater) # Arreglo con los índices de los máximos locales.
loc_min = argrelextrema(inter(x), np.less)    # Arreglo con los índices de los mínimos locales.

table_max = []
table_min = []

for i in loc_max[0]:
    if(0.<=x[i] and x[i]<=12.): # Especifico el intervalo en el que quiero los máximos y mínimos locales.
        table_max.append([x[i],float(inter(x[i]))]) # Genero la tabla con los máximos locales.
        
for i in loc_min[0]:
    if(0.<=x[i] and x[i]<=12.):
        table_min.append([x[i],float(inter(x[i]))]) # Genero la tabla con los mínimos locales.

headers = ["x values for local maximum","Interpolation function values"]        
print tabulate(table_max, headers,tablefmt="fancy_grid", numalign="center")
headers = ["x values for local maximum","Interpolation function values"]        
print tabulate(table_min, headers,tablefmt="fancy_grid", numalign="center")
```

La tabla con los datos teóricos es la siguiente:

```python
table_teo = [[0,1, 'Max'],[3.833,0,'Min'],[5.136,0.0175,'Max'],[7.016,0,'Min'],[8.417,0.0042,'Max'],[10.174,0,'Min'],[11.620,0.0016,'Max']]
headers = ["x values for local maximum and minimum","Function values for Airy Pattern", "Min or Max"]        
print tabulate(table_teo, headers,tablefmt="fancy_grid", numalign="center",stralign='center')
```

Al comparar los valores teóricos con los valores obtenidos se puede ver que coinciden muy bien. Claro está, la exactitud de las posicones y el valor de la función que describe el patrón de Airy están limitados por el número de puntos en en arreglo considerado, que este caso era 1024 (y de esos 1024, solo una fracción en el intervalo [0,12]). Entre más puntos de muestreo se tengan, mejor será la interpolación y, por ende, más cercanos estarán los valores experimentales de los teóricos en las tablas presentadas anteriormente.
