```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import math
%matplotlib inline
```

```python
angles = [math.radians(60*x) for x in range(6)]
sines = [math.sin(x) for x in angles]
cosin = [math.cos(x) for x in angles]

def L(angle, coords, jump):
    return (angle + 1) % 6
def R(angle, coords, jump):
    return (angle + 4) % 6
def F(angle, coords, jump):
    coords.append(
        (coords[-1][0] + jump * cosin[angle],
         coords[-1][1] + jump * sines[angle]))
    return angle

decode = dict(L=L, R=R, F=F)

def koch(order, length, origin):
    pathcodes="LFRFRF"
    for i in xrange(order):
        pathcodes = pathcodes.replace("F", "FLFRFLF")

    jump = float(length) / (3 ** order)
    coords = [origin]
    angle = 0

    for move in pathcodes:
        angle = decode[move](angle, coords, jump)
        

    return coords


TOTALWIDTH = 600.0

plt.figure(figsize=(15,15))
for i in range(0,6):
    points = koch(i,TOTALWIDTH,(0,10))
    plt.subplot(2,3,i+1,aspect='equal')
    plt.plot(np.asarray(points)[:,0],np.asarray(points)[:,1],color='red')
    plt.axis('off')
plt.show()
```

```python
%matplotlib

fig = plt.figure()

# ims is a list of lists, each row is a list of artists to draw in the
# current frame; here we are just animating one artist, the image, in
# each frame
ims = []
for i in range(6):
    points = koch(i,TOTALWIDTH,(0,10))
    im = plt.plot(np.asarray(points)[:,0],np.asarray(points)[:,1],color='black')
    ims.append(im)

ani = animation.ArtistAnimation(fig, ims, interval=500, blit=True,repeat_delay=500)

#ani.save('snowflake.gif',writer='imagemagick',fps=2,dpi=300)

plt.show()
```
