```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
alpha = 1.

def f(Y):
    return np.array([Y[1],-alpha*np.cos(Y[0])*Y[1]])

def one_Kutta4_step():
    global t
    global Deltat
    global Y
    K1 = f(Y)
    K2 = f(Y + Deltat*(1./2.*K1))
    K3 = f(Y + Deltat*(1./2.*K2))
    K4 = f(Y + Deltat*K3)
    Y+= Deltat * (1./6.*K1 + 1./3.*K2 + 1./3.*K3 + 1./6.*K4)
    t+= Deltat
```

```python
numIter = 10000 # Número de iteraciones a usar para solucionar las ecuaciones.

Y = np.array([0.,1.])

tmax = 10000. # Se especifica que se va a correr el método para media unidad de tiempo.

t = 0.

Deltat = tmax/numIter

history = np.array([[t,Y[0],Y[1]]])

for i in range(numIter): # Implementación del Runge Kutta para obtener los datos de la colución de las ecuaciones
    one_Kutta4_step()         # diferenciales.
    history = np.append(history,[[t,Y[0],Y[1]]],axis=0)
```

```python
time = history[:,0]
x1 = history[:,1]
plt.figure(figsize=(15,10))
plt.plot(time,x1)
plt.show()
```
