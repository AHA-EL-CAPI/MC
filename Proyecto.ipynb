```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
%matplotlib inline
```

```python
alpha = -15.
def f(Y,t):
    return np.array([Y[1],-alpha*np.cos(Y[0])*Y[1]]) 

times=np.linspace(0,0.1,10000)
funsol,others=odeint(f,[0.,1.],times,full_output=True)

plt.figure(figsize=(15,10))
plt.plot(times,funsol[:,0])
plt.show()
```

```python
alpha = 500.
def f(Y,t):
    return np.array([Y[1],-2*alpha*Y[0]*Y[1]]) 

times=np.linspace(0,.1,10000)
funsol,others=odeint(f,[0.,1.],times,full_output=True)

plt.figure(figsize=(15,10))
plt.plot(times,funsol[:,0])
plt.show()
```

```python
mass = 0.1
alpha = -15.
def f(Y,t):
    return np.array([Y[1],-alpha*np.cos(Y[0])*Y[1] + (1/mass)*np.cos(Y[0])*np.cos(t)]) 

times=np.linspace(0,50.,10000)
funsol,others=odeint(f,[0.,0.],times,full_output=True)

plt.figure(figsize=(15,10))
plt.plot(times,funsol[:,0])
plt.show()
```

```python
mass = 0.1
alpha = 15.
def f(Y,t):
    return np.array([Y[1],-alpha*np.cos(Y[0])*Y[1] + (1/mass)*np.cos(Y[0])*np.sin(t)]) 

times=np.linspace(0,50.,100000)
funsol,others=odeint(f,[0.,0.],times,full_output=True)

plt.figure(figsize=(15,10))
plt.plot(times,funsol[:,0])
plt.show()
```

```python
mass = 1.
alpha = 15.
def f(Y,t):
    return np.array([Y[1],-alpha*np.cos(Y[0])*Y[1] + (1/mass)*np.cos(Y[0])*np.exp(t)]) 

times=np.linspace(0,15.,100000)
funsol,others=odeint(f,[0.,0.],times,full_output=True)

plt.figure(figsize=(15,10))
plt.plot(times,funsol[:,0])
plt.show()
```

```python
mass = 1.
omega = 1.
alpha = .1
def f1(Y,t):
    return np.array([Y[1],-alpha*np.cos(Y[0])*Y[1] - (omega**2)*Y[0] + (1/mass)*np.cos(Y[0])*np.exp(t)]) 

def f2(Y,t):
    return np.array([Y[1],-alpha*Y[1] - (omega**2)*Y[0] + (1/mass)*np.cos(Y[0])*np.exp(t)]) 

times=np.linspace(0,15.,100000)
funsol1,others1=odeint(f1,[0.,1.],times,full_output=True)
funsol2,others2=odeint(f2,[0.,1.],times,full_output=True)

plt.figure(figsize=(15,10))
plt.plot(times,funsol1[:,0],'r')
plt.plot(times,funsol2[:,0],'b')
plt.show()
```
