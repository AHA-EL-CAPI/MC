```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def positions(N):
    positions = np.zeros((N,2)) # Array que guarda las posiciones x y y de los N puntos.
    for i in range(N):
        positions[i,0] = np.random.uniform(0.,1.)
        positions[i,1] = np.random.uniform(0.,1.) # Posiciones x y y de cada punto generadas aleatioriamente.
    return positions

def distances(Y):
    distances = np.array([]) # Array que guarda las distancias relevantes.
    for i in range(len(Y)):
        for j in range(len(Y)):
            if (j>i): # Condición necesaria para no calcular distancias redundantes o la distancia entre un punto
                      # y él mismo.
                distance = np.sqrt((Y[i,0]-Y[j,0])**2 + (Y[i,1]-Y[j,1])**2)
                distances = np.append(distances,distance)
    return distances
```

La idea es crear una array con las N posiciones de todos los puntos mediante la función `positions(N)` (con N el número de puntos a generar) y luego mirar las distancias no entre los mismos mediante la función `distances(Y)` (con Y el parámetro que contiene las posiciones de los puntos). Teniendo ahora el arreglo de distancias, la idea es  hallar el mínimo de ese array y repetir varias veces para un N fijo con el fin de obtener un promedio de distancia mínima para dicho N, todo lo anterior con el fin de cumplir con lo requerido en el parcial.

```python
def experiment(N,iteraciones):
    suma = 0. # Elemento necesario para calcular el promedio
    for i in range(iteraciones):
        a = positions(N)
        b = distances(a)
        suma += min(b) # En cada interación se crea el array de puntos, las distancias y se guarda el mínimo de las
                       # distancias.
    return suma/iteraciones # Promedio del mínimo de las distancias.
```

```python
x = [2,4,8,16,32,64,128]
y = [experiment(x[i],100) for i in range(len(x))]
fig = plt.figure(figsize=(15,10))
plt.grid('on')
ax = fig.add_subplot(1,1,1)
ax.plot(x,y,'o')
ax.set_xscale('log')
ax.set_yscale('log') # Gráfica logarítmica.
plt.xlabel(u'N',fontsize=20)
plt.ylabel(u'$d_{m}$',fontsize=20)
plt.title(u'Distancias mínimas promedio',fontsize=20)
plt.show()
```
