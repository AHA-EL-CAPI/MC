```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import ifft, fft, fftfreq
from scipy.io import wavfile
from scipy import interpolate
%matplotlib inline
```

```python
datos = np.genfromtxt("ring.dft",dtype='c16') # Importo los datos.
clean = ifft(datos).real # Hago la transformada inversa y tomo la parte real.
clean_1 = np.int16(clean) # Convierto a enteros el arreglo resultante de la transformada inversa.
wavfile.write('clean.wav',22050,clean_1) # Guardo en un archivo .wav el resultado.
plt.plot(clean) # Grafico para ver los datos de amplitudes.
plt.show()
```

```python
# Defino la función duplicar para aumentar la cantidad de datos de un array pasado como parámetro,
# interpolando con los puntos medios.
def duplicar(a):
    new_points = [(a[i+1]-a[i])/2. for i in range(len(a)-1)]
    b = [a[0]]
    for i in range(len(new_points)):
         b.append(a[i]+new_points[i])       
         b.append(a[i+1])   
    return b
```

```python
clean_2 = duplicar(clean) # Duplico los datos del arreglo limpio.
clean_3 = [clean_2[i]*(1. - float(i)/len(clean_2)) for i in range(len(clean_2))] # Aplico la rampa lineal.
clean_4 = np.int16(clean_2) # Convierto a enteros nuevamente.
wavfile.write('clean_rampa_lineal.wav',22050,clean_4) # Guardo en .wav
plt.plot(clean_3) # Grafico para verificar el efecto de la rampa.
plt.show()
```
