```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import ifft, fft, fftfreq
from scipy.io import wavfile
from scipy import interpolate
%matplotlib inline
```

```python
datos = np.genfromtxt("ring.dft",dtype='c16')
clean = ifft(datos).real
clean_1 = np.int16(clean)
wavfile.write('clean.wav',22050,clean_1)
plt.plot(clean)
```

```python
def duplicar(a):
    new_points = [(a[i+1]-a[i])/2. for i in range(len(a)-1)]
    b = [a[0]]
    for i in range(len(new_points)):
         b.append(a[i]+new_points[i])       
         b.append(a[i+1])   
    return b
```

```python
clean_2 = duplicar(clean)
clean_3 = [clean_2[i]*(1. - float(i/len(clean_2))) for i in range(len(clean_2))]
clean_4 = np.int16(clean_2)
wavfile.write('clean_4.wav',22050,clean_4)
plt.plot(clean_3)
```
