```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
%matplotlib inline
```

```python
x = np.array([0.0,np.pi/2,np.pi,3*np.pi/2,2*np.pi])
y = np.array(np.sin(x))

def lagrangep(absc,j):
    pol=np.poly1d([0,1])
    k=len(absc)
    for m in range(k):
        if m!=j:
            pol*=1./(absc[j]-absc[m])*np.poly1d([1.,-absc[m]])
    return pol 

def interlagr(absc,orde):
    poly=np.poly1d([0])
    for i in range(len(absc)):
        poly+=orde[i]*lagrangep(absc,i)
    return poly

po=interlagr(x,y)
fig=plt.figure()
ixv=np.linspace(-np.pi/2,5*np.pi/2,100)
ax=fig.add_subplot(111)
ax.plot(x,y,"o",ms=10,label="puntos muestrales")
ax.plot(ixv,po(ixv),label=u"Interpolación")
plt.legend(loc='best')
plt.show(fig)
```

```python
x_new = np.array(np.linspace(-np.pi/2, 2*np.pi, 15))
y_new = np.array(np.sin(x_new))

po1=interlagr(x_new,y_new)
fig=plt.figure()
ixv=np.linspace(-np.pi/2,2*np.pi,100)
ax=fig.add_subplot(111)
ax.plot(x_new,y_new,"o",ms=10,label="puntos muestrales")
ax.plot(ixv,po1(ixv),label=u"Interpolación")
plt.legend(loc='best')
plt.show(fig)
```
